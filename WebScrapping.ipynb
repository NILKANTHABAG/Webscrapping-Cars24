# Web Scrapping

1. `requests` - This is used to extract the HTML code from the given URL
2. `BeautifulSoup` - Format and Scrap the data from the HTML

**Steps**

1. Identify URL
2. Inspect HTML code
3. Find the HTML tag for the element that you want to extract.
4. Write some code to scrap this data

```python
# Loading required libraries

import numpy as np
import pandas as pd

import requests
from bs4 import BeautifulSoup
```

```python
# Identify the URL

URL = 'https://www.cars24.com/buy-used-car?sort=P&storeCityId=2&pinId=110001'
```

```python
# Loading the WebPage in Memory using requests library

page = requests.get(URL)
```

```python
# Check the Status Code of the Page

page.status_code
```

```python
# Extracting the HTML Code of the WebPage

htmlCode = page.text
```

```python
htmlCode
```

```python
# Format the HTML code using bs4 library

soup = BeautifulSoup(htmlCode,'lxml')
```

```python
print(soup.prettify())
```

```python
help(soup)
```

### find_all()

```python
names = soup.find_all('h2',attrs={'class' : '_3FpCg'})

for name in names:
     print(name.text)
```

```python
price = soup.find_all('div',attrs={'class' : '_7udZZ'})

for prize in price:
    print(prize.text)
```

```python
features = soup.find_all('ul',attrs={'class' : 'bVR0c'})

for feature in features:
    print(feature.text)
```

```python
emis=soup.find_all('strong')
    
for emi in emis:
    print(emi.text)
```

### Let's look into all the URLs

### Code for Web Scrapping (Correct way)

```python
# Scrapping the Web Page

title =[]
money=[]
attribute=[]
payment=[]
for i in range(1, 77):
    
    URL = "https://www.cars24.com/buy-used-car/?sort=P&page={} &storeCityId=2&pinId=110001".format(i)
        
            
    
    page = requests.get(URL)
    htmlCode = page.text
    
    soup = BeautifulSoup(htmlCode)
    
    for x in soup.find_all('div',attrs={'class' : '_7udZZ'}):
        name = soup.find('h2',attrs={'class' : '_3FpCg'})
        if name is None:
             title.append(np.NaN)
        else:
            title.append(name.text)
            
        
        price = soup.find('div',attrs={'class' : '_7udZZ'})
        if price is None:
             money.append(np.NaN)
        else:
             money.append(price.text)
                
        features = soup.find('ul',attrs={'class' : 'bVR0c'})
        if features is None:
             attribute.append(np.NaN)
        else:
             attribute.append(features.text)
                
                       
        emis=soup.find('strong')
        if features is None:
            payment.append(np.NaN)
        else:
             payment.append(emis.text)

       
```

```python
print(len(title))
print(len(money))
print(len(attribute))
print(len(payment))
```

# Create a DataFrame and save it in CSV file

```python
df = pd.DataFrame({'PRODUCT' : title,'DETAIL':attribute,'MRP':money,'EMI':payment})
```

```python
df.drop_duplicates(subset ="EMI",
                     keep = "first", inplace = True)
```

# Extracting the Details using Regex

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("cars.csv")

df.head()
```

```python
import regex as re
regex = '[\d]+[,]+[\d]+[\s]+km'

df['MILAGECOVERED'] = df['DETAIL'].apply(lambda x : re.findall(regex, x))
```

```python

import regex as re
regex = '[\d]+[a-z]+[a-z]+[\s]+Owner'

df['INHAND'] = df['DETAIL'].apply(lambda x : re.findall(regex, x))
```

```python
import regex as re
regex = 'Petrol|Diesel|Petrol+CNG}{1,1}'

df['ENGINETYPE'] = df['DETAIL'].apply(lambda x : re.findall(regex, x))
```

```python
import regex as re
regex = '[1-90]{4}'

df['MANUFACTURE'] = df['PRODUCT'].apply(lambda x : re.findall(regex, x))
```

```python
import regex as re
regex = '[^1-90]+[\d\w]'

df['NAME'] = df['PRODUCT'].apply(lambda x : re.findall(regex, x))
```
